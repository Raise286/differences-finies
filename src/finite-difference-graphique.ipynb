```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns
from typing import Callable, Tuple, List
import pandas as pd

# Configuration pour de beaux graphiques
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

class FiniteDifferenceSolver:
    """Solveur de différences finies pour -u'' = f avec conditions de Dirichlet"""
    
    def __init__(self, n: int, f: Callable[[float], float], u0: float, u1: float):
        self.n = n  # nombre de points intérieurs
        self.f = f  # fonction f dans -u'' = f
        self.u0 = u0  # u(0)
        self.u1 = u1  # u(1)
        self.h = 1.0 / (n + 1)  # pas de discrétisation
        
    def solve_direct(self) -> Tuple[np.ndarray, np.ndarray]:
        """Résolution directe par élimination de Gauss"""
        # Créer le second membre
        b = np.zeros(self.n)
        for i in range(self.n):
            xi = (i + 1) * self.h
            b[i] = self.h**2 * self.f(xi)
        
        # Conditions aux limites
        b[0] += self.u0
        b[self.n-1] += self.u1
        
        # Matrice tridiagonale: diagonale = 2, sur/sous-diagonales = -1
        diagonal = np.full(self.n, 2.0)
        upper_diag = np.full(self.n-1, -1.0)
        lower_diag = np.full(self.n-1, -1.0)
        
        # Résolution par algorithme de Thomas
        u_interior = self._solve_tridiagonal(lower_diag, diagonal, upper_diag, b)
        
        # Construire la solution complète
        u = np.zeros(self.n + 2)
        u[0] = self.u0
        u[1:-1] = u_interior
        u[-1] = self.u1
        
        # Points du maillage
        x = np.linspace(0, 1, self.n + 2)
        
        return x, u
    
    def solve_gauss_seidel(self, tol: float = 1e-10, max_iter: int = 10000, omega: float = 1.0) -> Tuple[np.ndarray, np.ndarray, int]:
        """Résolution par Gauss-Seidel avec relaxation"""
        # Initialisation
        u = np.zeros(self.n + 2)
        u[0] = self.u0
        u[-1] = self.u1
        
        # Interpolation linéaire initiale
        for i in range(1, self.n + 1):
            u[i] = self.u0 + (self.u1 - self.u0) * i * self.h
        
        iterations = 0
        error = float('inf')
        
        while error > tol and iterations < max_iter:
            max_change = 0.0
            
            for i in range(1, self.n + 1):
                old_value = u[i]
                xi = i * self.h
                new_value = 0.5 * (u[i-1] + u[i+1] - self.h**2 * self.f(xi))
                
                # Relaxation
                u[i] = (1 - omega) * old_value + omega * new_value
                
                change = abs(u[i] - old_value)
                max_change = max(max_change, change)
            
            error = max_change
            iterations += 1
        
        x = np.linspace(0, 1, self.n + 2)
        return x, u, iterations
    
    def _solve_tridiagonal(self, a: np.ndarray, b: np.ndarray, c: np.ndarray, d: np.ndarray) -> np.ndarray:
        """Algorithme de Thomas pour systèmes tridiagonaux"""
        n = len(d)
        x = np.zeros(n)
        
        # Forward sweep
        for i in range(1, n):
            factor = a[i-1] / b[i-1]
            b[i] -= factor * c[i-1]
            d[i] -= factor * d[i-1]
        
        # Back substitution
        x[n-1] = d[n-1] / b[n-1]
        for i in range(n-2, -1, -1):
            x[i] = (d[i] - c[i] * x[i+1]) / b[i]
        
        return x

def plot_comparison(x_mesh: np.ndarray, u_numerical: np.ndarray, 
                   exact_solution: Callable[[float], float], 
                   title: str, f_description: str) -> None:
    """Trace la comparaison entre solution numérique et exacte"""
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Solution continue pour la fonction exacte
    x_continuous = np.linspace(0, 1, 1000)
    y_exact_continuous = np.array([exact_solution(x) for x in x_continuous])
    y_exact_mesh = np.array([exact_solution(x) for x in x_mesh])
    
    # Graphique 1: Comparaison des solutions
    ax1.plot(x_continuous, y_exact_continuous, 'b-', linewidth=2, label='Solution exacte u(x)', alpha=0.8)
    ax1.plot(x_mesh, u_numerical, 'ro-', markersize=6, linewidth=2, label='Solution numérique', alpha=0.8)
    ax1.plot(x_mesh, y_exact_mesh, 'g^', markersize=8, label='u exacte aux points du maillage', alpha=0.7)
    
    ax1.set_xlabel('x')
    ax1.set_ylabel('u(x)')
    ax1.set_title(f'{title}\n{f_description}')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Graphique 2: Erreur ponctuelle
    error_pointwise = np.abs(u_numerical - y_exact_mesh)
    ax2.semilogy(x_mesh, error_pointwise, 'r*-', markersize=8, linewidth=2, label='Erreur |u_num - u_exact|')
    ax2.set_xlabel('x')
    ax2.set_ylabel('Erreur (échelle log)')
    ax2.set_title('Erreur ponctuelle')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def convergence_study(f: Callable[[float], float], exact_solution: Callable[[float], float],
                     u0: float, u1: float, title: str) -> None:
    """Étude de convergence avec différentes tailles de maillage"""
    
    n_values = [10, 20, 40, 80, 160]
    errors_l2 = []
    errors_max = []
    mesh_sizes = []
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    colors = plt.cm.viridis(np.linspace(0, 1, len(n_values)))
    
    for i, n in enumerate(n_values):
        solver = FiniteDifferenceSolver(n, f, u0, u1)
        x_mesh, u_num = solver.solve_direct()
        
        # Calcul des erreurs
        u_exact = np.array([exact_solution(x) for x in x_mesh])
        error_l2 = np.sqrt(np.trapz((u_num - u_exact)**2, x_mesh))
        error_max = np.max(np.abs(u_num - u_exact))
        
        errors_l2.append(error_l2)
        errors_max.append(error_max)
        mesh_sizes.append(1.0 / (n + 1))
        
        # Tracer quelques solutions pour visualiser la convergence
        if i % 2 == 0:  # Ne tracer que quelques courbes pour éviter l'encombrement
            ax1.plot(x_mesh, u_num, 'o-', color=colors[i], 
                    label=f'n={n}, h={1.0/(n+1):.3f}', alpha=0.7)
    
    # Solution exacte continue
    x_continuous = np.linspace(0, 1, 1000)
    y_exact = np.array([exact_solution(x) for x in x_continuous])
    ax1.plot(x_continuous, y_exact, 'k-', linewidth=3, label='Solution exacte', alpha=0.9)
    ax1.set_xlabel('x')
    ax1.set_ylabel('u(x)')
    ax1.set_title(f'{title} - Convergence du maillage')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Convergence en norme L2
    ax2.loglog(mesh_sizes, errors_l2, 'bo-', linewidth=2, markersize=8, label='Erreur L2')
    # Ligne de référence h²
    ax2.loglog(mesh_sizes, [0.5 * h**2 for h in mesh_sizes], 'r--', 
              linewidth=2, label='O(h²)', alpha=0.7)
    ax2.set_xlabel('Taille de maille h')
    ax2.set_ylabel('Erreur L2')
    ax2.set_title('Convergence en norme L2')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Convergence en norme max
    ax3.loglog(mesh_sizes, errors_max, 'ro-', linewidth=2, markersize=8, label='Erreur max')
    ax3.loglog(mesh_sizes, [0.1 * h**2 for h in mesh_sizes], 'g--', 
              linewidth=2, label='O(h²)', alpha=0.7)
    ax3.set_xlabel('Taille de maille h')
    ax3.set_ylabel('Erreur max')
    ax3.set_title('Convergence en norme infinie')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # Ordre de convergence
    orders_l2 = []
    orders_max = []
    for i in range(1, len(errors_l2)):
        if errors_l2[i] > 0 and errors_l2[i-1] > 0:
            order_l2 = np.log(errors_l2[i]/errors_l2[i-1]) / np.log(mesh_sizes[i]/mesh_sizes[i-1])
            orders_l2.append(order_l2)
        if errors_max[i] > 0 and errors_max[i-1] > 0:
            order_max = np.log(errors_max[i]/errors_max[i-1]) / np.log(mesh_sizes[i]/mesh_sizes[i-1])
            orders_max.append(order_max)
    
    ax4.plot(range(1, len(orders_l2)+1), orders_l2, 'bo-', linewidth=2, markersize=8, label='Ordre L2')
    ax4.plot(range(1, len(orders_max)+1), orders_max, 'ro-', linewidth=2, markersize=8, label='Ordre max')
    ax4.axhline(y=2, color='k', linestyle='--', alpha=0.7, label='Ordre théorique = 2')
    ax4.set_xlabel('Raffinement')
    ax4.set_ylabel('Ordre de convergence')
    ax4.set_title('Ordre de convergence numérique')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Afficher les résultats numériques
    print(f"\n=== Étude de convergence: {title} ===")
    print("n\t\th\t\tErreur L2\t\tErreur Max\t\tOrdre L2\tOrdre Max")
    print("-" * 80)
    for i in range(len(n_values)):
        order_l2_str = f"{orders_l2[i-1]:.2f}" if i > 0 and i-1 < len(orders_l2) else "-"
        order_max_str = f"{orders_max[i-1]:.2f}" if i > 0 and i-1 < len(orders_max) else "-"
        print(f"{n_values[i]}\t\t{mesh_sizes[i]:.4f}\t\t{errors_l2[i]:.2e}\t\t{errors_max[i]:.2e}\t\t{order_l2_str}\t\t{order_max_str}")

def compare_solvers(f: Callable[[float], float], exact_solution: Callable[[float], float],
                   u0: float, u1: float, n: int, title: str) -> None:
    """Compare les solveurs direct et Gauss-Seidel"""
    
    solver = FiniteDifferenceSolver(n, f, u0, u1)
    
    # Résolution directe
    x_mesh, u_direct = solver.solve_direct()
    
    # Résolution Gauss-Seidel
    x_mesh_gs, u_gs, iterations = solver.solve_gauss_seidel(tol=1e-10)
    
    # Solution exacte
    x_continuous = np.linspace(0, 1, 1000)
    y_exact = np.array([exact_solution(x) for x in x_continuous])
    y_exact_mesh = np.array([exact_solution(x) for x in x_mesh])
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Comparaison des solutions
    ax1.plot(x_continuous, y_exact, 'k-', linewidth=3, label='Solution exacte', alpha=0.9)
    ax1.plot(x_mesh, u_direct, 'bo-', markersize=6, linewidth=2, label='Solveur direct', alpha=0.8)
    ax1.plot(x_mesh_gs, u_gs, 'r^-', markersize=6, linewidth=2, label=f'Gauss-Seidel ({iterations} iter.)', alpha=0.8)
    ax1.set_xlabel('x')
    ax1.set_ylabel('u(x)')
    ax1.set_title(f'{title} - Comparaison des solveurs')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Différence entre les deux méthodes
    diff_solvers = np.abs(u_direct - u_gs)
    ax2.semilogy(x_mesh, diff_solvers, 'mo-', markersize=6, linewidth=2, 
                label='|u_direct - u_GS|')
    ax2.set_xlabel('x')
    ax2.set_ylabel('Différence (échelle log)')
    ax2.set_title('Différence entre solveurs')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Calcul des erreurs
    error_direct_l2 = np.sqrt(np.trapz((u_direct - y_exact_mesh)**2, x_mesh))
    error_gs_l2 = np.sqrt(np.trapz((u_gs - y_exact_mesh)**2, x_mesh))
    
    print(f"\n=== Comparaison des solveurs: {title} ===")
    print(f"Erreur L2 (Direct): {error_direct_l2:.2e}")
    print(f"Erreur L2 (Gauss-Seidel): {error_gs_l2:.2e}")
    print(f"Itérations Gauss-Seidel: {iterations}")
    print(f"Différence max entre solveurs: {np.max(diff_solvers):.2e}")

def main():
    """Fonction principale - Reproduit les exemples du code Java"""
    
    print("=== Visualisation des Solutions d'Équations Différentielles ===\n")
    
    # Exemple 1: -u'' = 4π²sin(2πx) avec u(0) = 0, u(1) = 0
    # Solution exacte: u(x) = sin(2πx)
    print("1. Exemple sinusoïdal")
    f1 = lambda x: 4 * np.pi**2 * np.sin(2 * np.pi * x)
    exact1 = lambda x: np.sin(2 * np.pi * x)
    
    solver1 = FiniteDifferenceSolver(50, f1, 0.0, 0.0)
    x1, u1 = solver1.solve_direct()
    
    plot_comparison(x1, u1, exact1, 
                   "Exemple 1: Fonction sinusoïdale", 
                   "f(x) = 4π²sin(2πx), u(0) = u(1) = 0")
    
    # Exemple 2: -u'' = 2 avec u(0) = 0, u(1) = 0
    # Solution exacte: u(x) = x(1-x)
    print("\n2. Exemple fonction constante")
    f2 = lambda x: 2.0
    exact2 = lambda x: x * (1 - x)
    
    solver2 = FiniteDifferenceSolver(30, f2, 0.0, 0.0)
    x2, u2 = solver2.solve_direct()
    
    plot_comparison(x2, u2, exact2,
                   "Exemple 2: Fonction constante",
                   "f(x) = 2, u(0) = u(1) = 0")
    
    # Exemple 3: -u'' = 6x avec u(0) = 0, u(1) = 0
    # Solution exacte: u(x) = x³ - x
    print("\n3. Exemple fonction linéaire")
    f3 = lambda x: 6 * x
    exact3 = lambda x: x**3 - x
    
    solver3 = FiniteDifferenceSolver(40, f3, 0.0, 0.0)
    x3, u3 = solver3.solve_direct()
    
    plot_comparison(x3, u3, exact3,
                   "Exemple 3: Fonction linéaire",
                   "f(x) = 6x, u(0) = u(1) = 0")
    
    # Exemple 4: Conditions aux limites non nulles
    # -u'' = 1 avec u(0) = 1, u(1) = 2
    print("\n4. Exemple avec conditions aux limites non nulles")
    f4 = lambda x: 1.0
    exact4 = lambda x: 1 + x + 0.5 * x * (1 - x)  # Solution approchée
    
    solver4 = FiniteDifferenceSolver(25, f4, 1.0, 2.0)
    x4, u4 = solver4.solve_direct()
    
    plot_comparison(x4, u4, exact4,
                   "Exemple 4: Conditions aux limites non nulles",
                   "f(x) = 1, u(0) = 1, u(1) = 2")
    
    # Études de convergence
    print("\n=== Études de convergence ===")
    
    convergence_study(f1, exact1, 0.0, 0.0, "Fonction sinusoïdale")
    convergence_study(f2, exact2, 0.0, 0.0, "Fonction constante")
    
    # Comparaison des solveurs
    print("\n=== Comparaison des solveurs ===")
    
    compare_solvers(f1, exact1, 0.0, 0.0, 50, "Fonction sinusoïdale")
    compare_solvers(f2, exact2, 0.0, 0.0, 40, "Fonction constante")

if __name__ == "__main__":
    main()
```
